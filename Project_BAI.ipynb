```python
from tkinter import *
from tkinter import messagebox

def get_hc():
        '''
           This function gets hip circumference value from Entry field
        '''
        hc = float(ENTRY1.get())
        return hc

def get_height():
        '''
           This function gets height value from Entry field
        '''
        height = float(ENTRY2.get())
        return height

def get_age():
        '''
           This function gets age value from Entry field
        '''
        age = int(ENTRY3.get())
        return age 


def get_gender():
        '''
           This function gets gender value from Entry field
        '''
        gender= (ENTRY4.get())
        return gender 
def clear(a=""):
            print(a)
            
            height = ENTRY2.delete(0,'end')
            hc = ENTRY1.delete(0,'end')
            age= ENTRY3.delete(0,'end')
            gender=(ENTRY4.delete(0,'end'))
            ENTRY1.focus_set()
            
def new_entry():
            MsgBox = messagebox.askquestion("New Entry","Do you want to Enter New Data?")
            if MsgBox == 'no':
                TOP.destroy()
            else:
                clear()
                messagebox.showinfo('Redirect','You will now redirected to the application screen')
                calculate_bai


def calculate_bai(a=""):   # "a" is there because the bind function gives an argument to the function....
        print(a)
        '''
          This function calculates the result
        '''
        try:
            height = get_height()
            hc = get_hc()
            age= get_age()
            gender=get_gender()
            height = height / 100.0
            bai = (hc / (height ** 1.5))-18
        except ZeroDivisionError:
            messagebox.showinfo("Result", "Please enter positive height!!")
        except ValueError:
            messagebox.showinfo("Result", "Please enter valid data!")
        else:
        
            if 20<=age<=39: 
                if (gender=='F'):
                    if bai<21.0:
                        res = "Your BAI is " + str(bai) + "\nRemarks: You are underweight!!"
                        messagebox.showinfo("Result", res)
                        new_entry()
                                       
                    elif 21<=bai<=33:
                        res = "Your BAI is " + str(bai) + "\nRemarks: You are Healthy!!"
                        messagebox.showinfo("Result", res)
                        new_entry()
                    
                    elif 33<=bai<=39:
                        res = "Your BAI is " + str(bai) + "\nRemarks: You are Overweight!!"
                        messagebox.showinfo("Result", res)
                        new_entry()
                    else:
                        res = "Your BAI is " + str(bai) + "\nRemarks: You are Obessed!!"
                        messagebox.showinfo("Result", res)
                        new_entry()
             
                elif(gender=='M'):
                    if bai<8.0:
                        res = "Your BAI is " + str(bai) + "\nRemarks: You are underweight!!"
                        messagebox.showinfo("Result", res)
                        new_entry()
                
                    elif 8<=bai<=21:
                        res = "Your BAI is " + str(bai) + "\nRemarks: You are Healthy!!"
                        messagebox.showinfo("Result", res)
                        new_entry()
                    
                    elif 21<=bai<=26:
                        res = "Your BAI is " + str(bai) + "\nRemarks: You are Overweight!!"
                        messagebox.showinfo("Result", res)
                        new_entry()
                    else:
                        res = "Your BAI is " + str(bai) + "\nRemarks: You are Obessed!!"
                        messagebox.showinfo("Result", res)
                        new_entry()

                else:
                     messagebox.showinfo("Result", "Please enter valid gender!")
                
        
            elif 40<=age<=59: 
                if (gender=='F'):
                    if bai<23.0:
                        res = "Your BAI is " + str(bai) + "\nRemarks: You are underweight!!"
                        messagebox.showinfo("Result", res)
                        new_entry()
                
                    elif 23<=bai<=35:
                        res = "Your BAI is " + str(bai) + "\nRemarks: You are Healthy!!"
                        messagebox.showinfo("Result", res)
                        new_entry()
                    
                    elif 35<=bai<=41:
                        res = "Your BAI is " + str(bai) + "\nRemarks: You are Overweight!!"
                        messagebox.showinfo("Result", res)
                        new_entry()
                    else:
                        res = "Your BAI is " + str(bai) + "\nRemarks: You are Obessed!!"
                        messagebox.showinfo("Result", res)
                        new_entry()
             
                elif(gender=='M'):
                    if bai<11.0:
                        res = "Your BAI is " + str(bai) + "\nRemarks: You are underweight!!"
                        messagebox.showinfo("Result", res)
                        new_entry()
                
                    elif 11<=bai<=23:
                        res = "Your BAI is " + str(bai) + "\nRemarks: You are Healthy!!"
                        messagebox.showinfo("Result", res)
                        new_entry()
                    
                    elif 23<=bai<=29:
                        res = "Your BAI is " + str(bai) + "\nRemarks: You are Overweight!!"
                        messagebox.showinfo("Result", res)
                        new_entry()
                    else:
                        res = "Your BAI is " + str(bai) + "\nRemarks: You are Obessed!!"
                        messagebox.showinfo("Result", res)
                        new_entry()
                
                else:
                     messagebox.showinfo("Result", "Please enter valid gender!")
                
            
            elif 60<=age<=79: 
                if (gender=='F'):
                    if bai<25.0:
                        res = "Your BAI is " + str(bai) + "\nRemarks: You are underweight!!"
                        messagebox.showinfo("Result", res)
                        new_entry()
                
                    elif 25<=bai<=38:
                        res = "Your BAI is " + str(bai) + "\nRemarks: You are Healthy!!"
                        messagebox.showinfo("Result", res)
                        new_entry()
                    
                    elif 38<=bai<=43:
                        res = "Your BAI is " + str(bai) + "\nRemarks: You are Overweight!!"
                        messagebox.showinfo("Result", res)
                        new_entry()
                    else:
                        res = "Your BAI is " + str(bai) + "\nRemarks: You are Obessed!!"
                        messagebox.showinfo("Result", res)
                        new_entry()
             
                elif(gender=='M'):
                    if bai<13.0:
                        res = "Your BAI is " + str(bai) + "\nRemarks: You are underweight!!"
                        messagebox.showinfo("Result", res)
                        new_entry()
                
                    elif 13<=bai<=25:
                        res = "Your BAI is " + str(bai) + "\nRemarks: You are Healthy!!"
                        messagebox.showinfo("Result", res)
                        new_entry()
                    
                    elif 25<=bai<=31:
                        res = "Your BAI is " + str(bai) + "\nRemarks: You are Overweight!!"
                        messagebox.showinfo("Result", res)
                        new_entry()
                    else:
                        res = "Your BAI is " + str(bai) + "\nRemarks: You are Obessed!!"
                        messagebox.showinfo("Result", res)
                        new_entry()
                
                else:
                     messagebox.showinfo("Result", "Please enter valid gender!")


if __name__ == '__main__':
    
            TOP = Tk()
            TOP.bind("<Return>", calculate_bai)
            TOP.geometry('500x500')
            TOP.configure(background="#307678")
            TOP.title("BAI Calculator")
            TOP.resizable(width=False, height=True)
            #Dispaly the information about BAI
            msg="The Body Adiposity Index (BAI) is a method of estimating the amount of body fat in humans. The BAI is calculated without using body weight, unlike the body mass index (BMI). Instead, it uses the size of the hips compared to the person's height"
            messagebox.showinfo("Welcome",str(msg))
            #Display the message on screen whether User wants to continue?
            MsgBox = messagebox.askquestion("BAI Calculator","Do you want to continue?")
            if MsgBox == 'no':
                TOP.destroy()
            else:
                messagebox.showinfo('Redirect','You will now redirected to the application screen')
        

            LABLE = Label(TOP, bg="#307678", text="Welcome to BAI Calculator", font=("Helvetica", 15, "bold"), pady=10)
            LABLE.place(x=55, y=0)

            LABLE1 = Label(TOP, bg="#cef0f1", text="Hip_Circumference(in cm)", bd=6,
                       font=("Helvetica", 8, "bold"), pady=5)
            LABLE1.place(x=55, y=60)
            ENTRY1 = Entry(TOP, bd=8, width=6, font="Roboto 11")
            ENTRY1.place(x=240, y=60)

            LABLE2 = Label(TOP, bg="#cef0f1", text="Height(in cm)", bd=6,
                       font=("Helvetica", 8, "bold"), pady=5)
            LABLE2.place(x=55, y=121)
            ENTRY2 = Entry(TOP, bd=8, width=6, font="Roboto 11")
            ENTRY2.place(x=240, y=121)

            LABLE3 = Label(TOP, bg="#cef0f1", text="Age :", bd=6,
                       font=("Helvetica", 8, "bold"), pady=5)
            LABLE3.place(x=55, y=180)
            ENTRY3 = Entry(TOP, bd=8, width=6, font="Roboto 11")
            ENTRY3.place(x=240, y=180)

            LABLE4 = Label(TOP, bg="#cef0f1", text="Gender (F/M) :", bd=6,
                       font=("Helvetica", 8, "bold"), pady=5)
            LABLE4.place(x=55, y=240)
            ENTRY4 = Entry(TOP, bd=8, width=6, font="Roboto 11")
            ENTRY4.place(x=240, y=240)


            BUTTON =Button(bg="#2187e7", bd=12, text="BAI", command=calculate_bai,
                        font=("Helvetica", 10, "bold"))
            BUTTON.grid(row=6, column=0, sticky=W)
            BUTTON.place(x=115, y=360)
            TOP.mainloop()
        
    
    
    
    
 
```
